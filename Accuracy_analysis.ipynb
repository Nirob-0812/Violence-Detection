```python
import matplotlib.pyplot as plt

model_names = ['MobileNet-V2+LSTM','VGG19', 'VGG16', 'ResNet50' , "CNN+LSTM"]
accuracy = [95.0, 83.0, 87.0,70.0,98.0]

plt.figure(figsize=(10, 5))
bars = plt.bar(model_names, accuracy, color='skyblue')

# Adding data labels
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval - 3, round(yval, 2), ha='center', va='bottom')

plt.xlabel("Models", fontsize=12)
plt.ylabel("Accuracy (%)", fontsize=12)
plt.title("Comparison of Model's Accuracy", fontsize=14)
plt.ylim([0, 100])  # Limiting y-axis to 100
plt.grid(True)  # Adding grid
plt.show()
```
